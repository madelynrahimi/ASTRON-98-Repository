```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,3,100)
yerr = 0.05 + 2*x

```

```python
plt.figure()
plt.plot(x,x**2, label = 'x^2')
plt.plot(x,x**3, label = 'x^3')
plt.plot(x,x**4, label = 'x^4')
plt.plot(x,x**5, label = 'x^5')

plt.title('exponential plots')
plt.xlabel('x values')
plt.ylabel('exponents')
plt.legend()
plt.show()
```

```python
plt.figure()
for p in range(2,6):
    plt.plot(x, x**p, label = 'x^' + str(p))
    plt.errorbar(x,x**p,yerr)
plt.legend()
plt.show()
```

```python
x = np.linspace(0,5,100)

y1 = np.exp(x)
y2= np.exp(2*x)
y3 = np.exp(x)+1
y4 = np.exp(x)-1

fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y1)
axs[0,1].plot(x,y2)
axs[1,0].plot(x,y3)
axs[1,1].plot(x,y4)
plt.tight_layout()
plt.show()
```
