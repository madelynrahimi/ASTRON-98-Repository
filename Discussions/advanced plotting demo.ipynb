```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
x,y = np.meshgrid(np.linspace(-1,1,100),np.linspace(-2,2,100))
z = x + y 

plt.figure
plt.imshow(z, cmap = "inferno")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar()
plt.show
```

```python
x,y = np.meshgrid(np.linspace(-5,5,100),np.linspace(-5,5,100))
z = ((x**2) + (y**2))**0.5

plt.figure
plt.imshow(z, cmap = "cool")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar()
plt.show
```
