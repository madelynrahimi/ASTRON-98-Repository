```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P,V = np.meshgrid(np.linspace(1,10,100),np.linspace(10,30,100))
T = ((P + (16.02/V**2))*(V - 0.1124))/0.083144

plt.figure
plt.imshow(T, cmap = "inferno", vmin = 1500, vmax =1500)
plt.xlabel("Pressure")
plt.ylabel("Volume")
plt.title("Temperature of One Mole of Acetone")
plt.colorbar()
plt.show()
```
