```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculatePi(N):
    insideCircle = 0
    insideSquare = 0
    for i in range(int(N)):
        x = np.random.uniform(-1,1)
        y = np.random.uniform(-1,1)
        originDistance = np.sqrt(x**2 +y**2)
        if originDistance <= 1:
            insideCircle += 1  
        insideSquare += 1 
    pi = 4 * insideCircle/insideSquare
    return pi

print(calculatePi(1e3))
print(calculatePi(1e4))
print(calculatePi(1e5))
print(calculatePi(1e6))
```

```python
plt.figure
x1 = np.random.uniform(-1,1, size = int(1e4))
y1= np.random.uniform(-1,1, size = int(1e4))
t = np.linspace(0,np.pi*2,100)
x2= np.cos(t)
y2= np.sin(t)
plt.scatter(x1,y1)
plt.plot(x2,y2, color = 'red')
plt.show()

```
