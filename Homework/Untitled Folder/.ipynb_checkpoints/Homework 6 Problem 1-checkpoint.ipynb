```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.linspace(0,40,40)
val1 = np.random.randint(0,101, size = 40)
val2 = np.random.randint(0,101, size = 40)
val3 = np.random.randint(0,101, size = 40)

fig, axs = plt.subplots(2,1)
axs[0].plot(x,val1, color = 'orange', linewidth = '10')
axs[0].plot(x,val2, color = 'red', linestyle = 'dashed')
axs[0].set_title('Problem 1')
axs[0].set_xlabel('X Values')
axs[0].set_ylabel('Y Values')
axs[1].scatter(x,val3, marker = 'D', color = 'magenta')
axs[1].set_title('Problem 2')
axs[1].set_xlabel('X Values')
axs[1].set_ylabel('Y Values')

plt.suptitle('Homework 7 Problem 1 Graphs')
plt.tight_layout()
plt.show
```
