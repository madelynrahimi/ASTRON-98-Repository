```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
planets_ds = sns.load_dataset('planets')
planets_ds['orbital_period_new'] = np.log10(planets_ds['orbital_period'])
print(planets_ds)
```

```python
sns.scatterplot(data=planets_ds, x = ('orbital_period_new'), y = 'mass', hue = 'method').set(title='Logarithmic Orbital Period vs. Mass of Exoplanets Discovered')
plt.xlabel('Logarithmic Orbital Periods')
plt.ylabel('Exoplanet Mass')
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(data=planets_ds, x = ('year'), y = 'number', hue = 'method').set(title='Number of Exoplanets Discovered by Year')
plt.xlabel('Year')
plt.ylabel('Number of Exoplanets Discovered')
plt.show()
```
