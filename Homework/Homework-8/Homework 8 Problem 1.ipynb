```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as fit
```

```python
one_df_temp = pd.read_csv("GlobalLandTemperaturesByState.csv")
print(one_df_temp)
```

```python
two_df_temp = one_df_temp.loc[:,['dt','AverageTemperature','State']]
print(two_df_temp)
```

```python
three_df_temp = two_df_temp.loc[(two_df_temp['dt'].str.contains('200'))|(two_df_temp['dt'].str.contains('201'))]
print(three_df_temp)
```

```python
four_df_temp = three_df_temp.loc[(three_df_temp['State'].str.contains("Wyoming"))|(three_df_temp['State'].str.contains("Nebraska"))|(three_df_temp['State'].str.contains("South Dakota"))]
print(four_df_temp)
```

```python
w_df_temp = three_df_temp.loc[(three_df_temp['State'].str.contains("Wyoming"))]
wfinal_df_temp = w_df_temp.iloc[:, 0:2]
print(wfinal_df_temp)

n_df_temp = three_df_temp.loc[(three_df_temp['State'].str.contains("Nebraska"))]
nfinal_df_temp = n_df_temp.iloc[:, 0:2]
print(nfinal_df_temp)

sd_df_temp = three_df_temp.loc[(three_df_temp['State'].str.contains("South Dakota"))]
sdfinal_df_temp = sd_df_temp.iloc[:, 0:2]
print(sdfinal_df_temp)
```

```python
x = wfinal_df_temp.iloc[:,0]
y = wfinal_df_temp.iloc[:,1]
plt.figure()
plt.plot(x,y)
plt.title('Date vs. Average Temperature in Wyoming')
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.show()
```

```python
x = nfinal_df_temp.iloc[:,0]
y = nfinal_df_temp.iloc[:,1]
plt.figure()
plt.plot(x,y)
plt.title('Date vs. Average Temperature in Nebraska')
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.show()
```

```python
x = sdfinal_df_temp.iloc[:,0]
y = sdfinal_df_temp.iloc[:,1]
plt.figure()
plt.plot(x,y)
plt.title('Date vs. Average Temperature in South Dakota')
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.show()
```

```python
sdfinal_df_temp['dt']=sdfinal_df_temp['dt'].str.replace('-','')
sdfinal_df_temp['dt']=sdfinal_df_temp['dt'].astype(int)
print(sdfinal_df_temp)
```

```python
x_data = sdfinal_df_temp.iloc[:,0]
y_data = sdfinal_df_temp.iloc[:,1]

def f(x, a, b, c, d):
    return a* np.sin(b*x +c)+d

p0 = [20, np.pi, np.pi, 5]
```

```python
parameters, covariance_matrix = fit.curve_fit(f,x_data,y_data, p0)
parameters 
```

```python
a = parameters[0]
b = parameters[1]
c = parameters[2]
d = parameters[3]

x = np.linspace(x_data.min(),x_data.max(), 100)
y = f(x,a,b,c,d)
plt.figure()
plt.plot(x,y)
plt.scatter(x_data, y_data)
plt.show()
```

```python
a_error = covariance_matrix[0,0] ** 0.5
b_error = covariance_matrix[1,1] ** 0.5
c_error = covariance_matrix[2,2] ** 0.5
d_error = covariance_matrix[3,3] ** 0.5
```

```python
print('a={0:1.5f} +\- {1:1.5f}'.format(a,a_error))
print('b={0:1.5f} +\- {1:1.5f}'.format(b,b_error))
print('c={0:1.5f} +\- {1:1.5f}'.format(c,c_error))
print('d={0:1.5f} +\- {1:1.5f}'.format(d,d_error))
```
