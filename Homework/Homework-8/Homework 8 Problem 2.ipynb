```python
from astropy.table import Table 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as fit
```

```python
astro_ds = 'global_SF6_MM.dat'
read_astro_ds = Table.read(astro_ds, format='ascii')
print(read_astro_ds)
```

```python
initial_df_astro = read_astro_ds.to_pandas()
print(initial_df_astro)
```

```python
df_iastro = initial_df_astro.iloc[:,[0,6,7]]
df_astro = df_iastro.dropna()
print(df_astro)
```

```python
x_data = df_astro.columns[0]
y_data = df_astro.columns[1]
y_error = df_astro.columns[2]
df_astro.plot(kind = "scatter", x = x_data , y = y_data, yerr = y_error, title = "Year versus Global Mean Concentration")
plt.xlabel("Year")
plt.ylabel("Global Mean Concentration")
plt.show()
```

```python
x_val = df_astro.iloc[:, 0]
y_val = df_astro.iloc[:, 1]
y_err = df_astro.iloc[:, 2]
def f(x, m, b):
    return (m * x) +b
p0=[0.014,2]
parameters, covariance_matrix = fit.curve_fit(f,x_val,y_val,p0)
```

```python
parameters
```

```python
x = np.linspace(x_val.min(), x_val.max(), 217)
m = parameters[0]
b = parameters[1]
y = f(x, m, b)
df_astro.plot(kind = "scatter", x = x_data , y = y_data, yerr = y_error, title = "Year versus Global Mean Concentration")
plt.plot(x,y)
plt.xlabel("Year")
plt.ylabel("Global Mean Concentration")
plt.show()
```

```python
m_error = covariance_matrix[0,0]**0.5
b_error = covariance_matrix[1,1]**0.5
print('m={0:1.5f} +\- {1:1.5f}'.format(m,m_error))
print('b={0:1.5f} +\- {1:1.5f}'.format(b,b_error))
```

```python
df_astro = df_astro[df_astro['SF6ottoGLsd'] != 0.00]
print(df_astro)
```
